```python
def inirandland(L):
    land = np.random.randint(2,size=(L,L))
    return land
```

```python
def inicaminante(L):
    land = np.zeros([L,L])
    center = int(L/2)
    land[center-1][center] = 1
    land[center][center+1] = 1
    land[center+1][center] = 1
    land[center+1][center-1] = 1
    land[center+1][center+1] = 1
    return land
```

```python
def Linearpoten(i,j,c):
    '''i y j determinan la celda'''
    return c*(i+j)
```

```python
def Mapapotencial(L,M,c):
    land = np.zeros([L,M])
    for i in range(0,L):
        for j in range(0,M):
            land[i][j] = Linearpoten(i,j,c)
    return land
```

```python
def funcionextraer(lista):
    nummax = max(lista)
    for i in range(0,len(lista)):
        if lista[i]==nummax:
            indx = i
    return indx
```
